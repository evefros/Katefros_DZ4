```python
# Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.
# Общие требования к программе:

# код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
# в коде отсутствуют глобальные переменные (за исключением documents и directories);
# пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру

def people(doc_number):
    for i in documents:
        if i["number"] == doc_number:
            return f'Владелец документа: {i["name"]}'
            break
    else:
        return 'Документ не найден в базе'


# Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится

def shelf(doc_number):
 
    if doc_number not in sum(directories.values(), []):
        return 'Документ не найден'
    else:
        for key, value in directories.items():
            if doc_number in value:
                return(f'Документ хранится на полке №{key}')   

            
# Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам            

def list_all():
    for i in documents:
        print(f'Тип документа: {i["type"]}, №: {i["number"]}, владелец: {i["name"]}, {shelf(i["number"])}')

        
# Пункт 4. Пользователь по команде "ads" может добавить новую полку       

def add_shelf(shelf_number):
    if shelf_number in directories.keys():
        return f'Такая полка уже существует! {all_shelves()}'
    else:
        while shelf_number != '' and shelf_number != ' ':
            new_directories = {shelf_number: []}
            directories.update(new_directories)
            return f'Добавлена полка №{shelf_number}. {all_shelves()}'
        else:
            return 'Номер полки не может быть пустым значением!'
      
    
# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def del_shelf(shelf_number):
    if shelf_number not in directories.keys():
        return 'Такой полки не существует! {all_shelves()}'
    elif shelf_number in directories.keys() and directories[shelf_number] == []:
        del(directories[shelf_number])
        return f'Полка №{shelf_number} удалена. {all_shelves()}'
    else:
        return f'На полке есть документы, удалите их перед удалением полки. {all_shelves()}'

    
# Печать перечня полок для п.п.4, 5   

def all_shelves():
    for i in directories.keys():
        shelves = ', '.join(directories)
        return f'Текущий перечень полок {shelves}'
       
        
# Главное меню пользователя   

def main_menu():
    while True:
        command = input(f'Введите команду из списка:\n p - узнать имя владельца;\n s - номер полки хранения;\n l - полный список хранилица;\n ads - добавить новую полку;\n ds - удалить пустую полку;\n q - выход. \n\n\nВведите команду: ')
        if command == 'p':
            doc_number = input('Введите номер документа: ')
            print(people(doc_number))
        elif command == 's':
            doc_number = input('Введите номер документа: ')
            print(shelf(doc_number))
        elif command == 'l':
            return list_all()
            print(list_all())
        elif command == 'ads':
            print(add_shelf(input('Введите номер новой полки: ')))
        elif command == 'ds':
            print(del_shelf(input('Введите номер полки для удаления: ')))
        elif command == 'q':
            print('Спасибо за работу!')
            break
        else:
            print(f'Некорректная команда! Введите команду из списка:\n p - узнать имя владельца;\n s - номер полки хранения;\n l - полный список хранилица;\n ads - добавить новую полку;\n ds - удалить пустую полку;\n q - выход')
main_menu()   
```
